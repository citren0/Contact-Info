```python
import csv
import string
import requests
from bs4 import BeautifulSoup

input_dataset = csv.DictReader(open("Core_Dataset.csv"))

with open('OUTPUT_ContactInfo.csv', 'w', newline='') as f:
    writer = csv.writer(f, delimiter=',', quotechar='"', quoting=csv.QUOTE_MINIMAL)

    writer.writerow(['ContactID', 'OrganizationName', 'URLUsed', 'ContactType', 'ContactHandle'])
    contactHandle = ""
    
    for row in input_dataset:
        print (row)
        orgID = row['Contact ID']
        orgName = row['Organization Name']
        orgURL = row['Website']
        
        
        if orgID:
            # Go out to the organization's website and grab stuff
            try:
                page = requests.get(orgURL)     #to extract page from website
                html_code = page.content        #to extract html code from page
            except Exception as e:
                print(e)
                
            # Get all the HTML into soup to search by
            try:
                soup = BeautifulSoup(html_code, 'html.parser')  #Parse html code
            except Exception as e:
                print(e)

                
            # Go through all the soup to find the different social media and contact references available
            # There are 5 types of links we need to collect. There will be a collumn in the CSV file for the
            # type of social media and then another for the link to this social media.
            # I use 5 if statements to determine what kind of social media each link is for.
            
            
            for link in soup.find_all(href=re.compile("twitter")):
                if link.get('href').strip() != contactHandle: # Trying to prevent duplicate links which were very, very prevalent.
                    contactHandle = link.get('href').strip()
                    if contactHandle:
                        contactType = "Twitter"           
                        keep = True
                    
                        #Some links have strange javascript code in them. These links will not work
                        if "javascript" in contactHandle:
                            keep = False
                            
                        
                        
                        if contactHandle.find("/") == 0: 
                            #Don't keep invalid links that look like //facebook or //facebook.com/whogivesascrap which ive found dont work.
                            tempHandle = contactHandle
                            httpsVar = "http:"
                            contactHandle = httpsVar + contactHandle
                        
                        
                        print('Contact Handle: ', contactHandle)
                        print('Contact Type: ', contactType)
                        if keep:    
                            try:
                                # Writing what was found to the our CSV
                                writer.writerow([orgID, orgName, orgURL, contactType,contactHandle])
                            except Exception as e:
                                print(e)
                        
                        
                        
                        
                        
            for link in soup.find_all(href=re.compile("facebook")):
                if link.get('href').strip() != contactHandle: # Trying to prevent duplicate links which were very, very prevalent.
                    contactHandle = link.get('href').strip()
                    if contactHandle:
                        contactType = "Facebook"           
                        keep = True
                        
                        #Some links have strange javascript code in them. These links will not work
                        if "javascript" in contactHandle:
                            keep = False
                            
                        
                    
                        if contactHandle.find("/") == 0: 
                            #Fix invalid links that look like //facebook or //facebook.com/whogivesascrap which ive found dont work.
                            tempHandle = contactHandle
                            httpsVar = "http:"
                            contactHandle = httpsVar + contactHandle
                        
                        print('Contact Handle: ', contactHandle)
                        print('Contact Type: ', contactType)
                        if keep:    
                            try:
                                # Writing what was found to the our CSV
                                writer.writerow([orgID, orgName, orgURL, contactType,contactHandle])
                            except Exception as e:
                                print(e)
                        
            for link in soup.find_all(href=re.compile("linkedin")):
                if link.get('href').strip() != contactHandle: # Trying to prevent duplicate links which were very, very prevalent.
                    contactHandle = link.get('href').strip()
                    if contactHandle:
                        contactType = "LinkedIn"           
                        keep = True
                        
                        #Some links have strange javascript code in them. These links will not work
                        if "javascript" in contactHandle:
                            keep = False
                            
                        
                    
                        if contactHandle.find("/") == 0: 
                            #Don't keep invalid links that look like //facebook or //facebook.com/whogivesascrap which ive found dont work.
                            tempHandle = contactHandle
                            httpsVar = "http:"
                            contactHandle = httpsVar + contactHandle
                        
                        print('Contact Handle: ', contactHandle)
                        print('Contact Type: ', contactType)
                        if keep:    
                            try:
                                # Writing what was found to the our CSV
                                writer.writerow([orgID, orgName, orgURL, contactType,contactHandle])
                            except Exception as e:
                                print(e)
                        
            for link in soup.find_all(href=re.compile("googleplus")):
                if link.get('href').strip() != contactHandle: # Trying to prevent duplicate links which were very, very prevalent.
                    contactHandle = link.get('href').strip()
                    if contactHandle:
                        contactType = "Google Plus"           
                        keep = True
                        
                        #Some links have strange javascript code in them. These links will not work
                        if "javascript" in contactHandle:
                            keep = False
                            
                        
                    
                        if contactHandle.find("/") == 0: 
                            #Don't keep invalid links that look like //facebook or //facebook.com/whogivesascrap which ive found dont work.
                            tempHandle = contactHandle
                            httpsVar = "http:"
                            contactHandle = httpsVar + contactHandle
                        
                        print('Contact Handle: ', contactHandle)
                        print('Contact Type: ', contactType)
                        if keep:    
                            try:
                                # Writing what was found to the our CSV
                                writer.writerow([orgID, orgName, orgURL, contactType,contactHandle])
                            except Exception as e:
                                print(e)
            
            # Collecting the links for email is a little more difficult than all the others.
            for link in soup.find_all(href=re.compile("mail")):
                # Mail links come in the href tag and are preceded by "mailto:" and then the address
                contactHandle = link.get('href').strip()
                if "mailto:" in contactHandle:
                    # Removing the preceding mailto tag
                    contactHandle.replace('mail','')
                    contactType = "Email"           
                    keep = True
                    
                    #fix invalids like //facebook or //facebook.com/whogivesascrap which ive found dont work.
                    contactHandle.replace("/", "")
                        
                    #INCOMPLETE INCOMPLETE
                    #Some emails had strange endings and beginnings with % signs and such. This is made to remove them.
                    if "%" in contactHandle:
                        keep = False
                        
                    #Some links have strange javascript code in them. These links will not work
                    if "javascript" in contactHandle:
                        keep = False
                                
                    print('Contact Handle: ', contactHandle)
                    print('Contact Type: ', contactType)
                    if keep:    
                        try:
                            # Writing what was found to the our CSV
                            writer.writerow([orgID, orgName, orgURL, contactType,contactHandle.replace("mailto:","")])
                        except Exception as e:
                            print(e)

```
